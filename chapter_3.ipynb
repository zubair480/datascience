```python
import pandas as pd
import seaborn as sms
import matplotlib.pyplot as plt
import ssl

ssl._create_default_https_context = ssl._create_unverified_context

url = "https://www.murach.com/python_analysis/mortality.csv"
mortality_data = pd.read_csv(url)
print(mortality_data.head())

mortality_data.columns = mortality_data.columns.str.replace(' ', '')

mortality_wide = mortality_data.pivot(index='Year', columns='AgeGroup', values='DeathRate')
```

```python
mortality_data.head()
```

```python
mortality_data.query('AgeGroup == "1-4 Years"').plot.scatter(x='Year', y='DeathRate')
```

```python
mortality_wide.plot()
```

```python
print("Mortality Wide Plot Area")
mortality_wide.plot.area()
print("Mortality Wide Plot Bar")
mortality_wide.plot.bar()
print("Mortality Wide Plot Scatter")
mortality_data.plot.scatter(x='Year', y='DeathRate')
print("Mortality Wide Plot Horizontal Bar")
mortality_wide.query("Year in (1900,2000)").plot.bar()

mortality_wide.query("Year in (1900,2000)").plot.barh()
```

```python
mortality_data.plot.hist(y='DeathRate', bins=8)

mortality_data.plot.density(y='DeathRate')
```

```python
mortality_wide.plot.box()


mortality_data.groupby('AgeGroup')['DeathRate'].sum().plot.pie()
```
