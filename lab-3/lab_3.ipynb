```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Housing.csv') 
data.head() 
```

```python
import matplotlib.pyplot as plt

plt.xlabel('Price')
plt.ylabel('Area')
scatterplot = plt.scatter(data['price'], data['area'], color='blue', marker='+')
```

```python
from sklearn import linear_model
model_reg = linear_model.LinearRegression()
x = data[['area']]
y = data['price']
model_reg.fit(x, y)
print("Intercept:", model_reg.intercept_)
print("Coefficient:", model_reg.coef_)
```

```python
model_reg.predict([[3000]])
```

```python
print(data[['area','price']].info())
print(data[['area','price']].describe())
print(data[['area','price']].isna().sum())
print(data[['area','price']].head(10))
```

```python
plt.xlabel('Price')
plt.ylabel('Area')
plt.scatter(data['price'], data['area'], color='blue', marker='+')
plt.plot(data['area'], model_reg.predict(data[['area']]), color='red')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

X = data[['area']]
y = data['price']

from sklearn.linear_model import LinearRegression
model = LinearRegression().fit(X, y)

order = np.argsort(X.values.flatten())
plt.scatter(data['area'], data['price'], color='blue', marker='+')
plt.plot(X.values.flatten()[order], model.predict(X)[order], color='red')
plt.xlabel('Area')
plt.ylabel('Price')
plt.show()

print("R:", model.score(X, y))
print("Intercept:", model.intercept_)
print("Coefficient:", model.coef_)
```
